## 自訂相關函數

```python
def showrule(model_list):
    for item in model_list:
        pair = item[0]
        items = [x for x in pair]
        if len(items) == 2:
            print("Rule: " + items[0] + " --> " + items[1])
            print("Support: "+ " "*9 + "%-.7s" % (item[1]))
            print("Confidence: "+ " "*6  + "%-.7s" % (str(item[2][0][2])))
            print("Lift: "+ " "*12 + "%-.7s" % (str(item[2][0][3])))
            print('\n' + "="*60 + '\n')
        else:
            print("Rule: " + items[0])
            print("Support: "+ " "*9 + "%-.7s" % (item[1]))
            print('\n' + "="*60 + '\n')
```

### --> 關聯規則的呈現公式

```python
def draw_bar_chart(qty,prd,xlabel,title):
    labels = np.array(prd)
    colors = ['#008000','#808000','#FFFF00','#000000','#FF0000','#00FF00','#0000FF','#008080','#aa22ff','#aa22ff','#dd0022','#ff00cc','#eeaa22','#22bbaa','#C0C0C0']
    y_pos = np.arange(len(labels))
    x_pos = np.array(qty)
    plt.barh(y_pos, x_pos, color = colors, align = 'center', edgecolor = 'green')
    plt.yticks(y_pos, labels, fontsize = 24)
    plt.xticks(fontsize = 24)
    plt.ylabel("貨品名稱", fontsize = 36)
    plt.xlabel(xlabel, fontsize = 36)
    plt.title((title + "\n"), fontsize = 36)
    plt.show()
```

# 然後將應用關聯規則(Association Rules)分析，試著找出更多潛在關鍵商品 

```python
df_list = []   #已知天爸水果為關鍵產品，故先剔除(使用df2)
df_group = df2.groupby(["單據日期","客戶簡稱"])
for item in df_group:
    df_list.append(item[1]["貨品名稱"].tolist())
```

```python
model_list = list(apriori(transactions = df_list, min_support = 0.0025, min_confidence = 0.03, min_left = 1.2, max_length = 2))
```

```python
print("規則總數：" + str(len(model_list)) + "條")
```

## 以支持度(Support)進行排序，並分別以資訊及圖顯示出規則

```python
support_rank = sorted(model_list, reverse = True, key = lambda RelationRecord : RelationRecord.support)
showrule(support_rank)
```

```python
plt.style.use('ggplot')
plt.figure(figsize = (40,30))
model_df_sup = pd.DataFrame(support_rank)
support = model_df_sup.iloc[0:19]['support']
products = model_df_sup.iloc[0:19]['items']
draw_bar_chart(support, products,"支持度","消費者支持度排名")
```

## 以信賴度(Confidence)進行排序，並以資訊顯示出規則

```python
confidence_rank = sorted(model_list, reverse = True, key = lambda RelationRecord : RelationRecord.ordered_statistics[0][2])
showrule(confidence_rank)
```

## 以提升值(Lift)進行排序，並以資訊顯示出規則

```python
lift_rank = sorted(model_list, reverse = True, key = lambda RelationRecord : RelationRecord.ordered_statistics[0][3])
showrule(lift_rank)
```

# 為了方便檢視，以網狀圖呈現關聯規則所導出之關鍵產品落點

```python
things = []
arrow = []
for item in model_list:
    pair = item[0]
    items = [x for x in pair]
    if len(items) == 2:
        if items[0] not in things:
            things.append(items[0])
        if items[1] not in things:
            things.append(items[1])
        arrow.append((items[0],items[1]))

plt.figure(figsize = (15,15))
G = nx.DiGraph()

G.add_nodes_from(things)
G.add_edges_from(arrow)
nx.draw(G, with_labels = True, node_size = 5000, node_color = 'yellow')
plt.show()
```
