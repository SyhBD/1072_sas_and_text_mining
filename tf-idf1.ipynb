```python
import jieba
import sys

from sklearn import feature_extraction
from sklearn.feature_extraction.text import TfidfVectorizer

from jieba import analyse
```

```python
#每篇文章的每個字詞的tf-idf值
corpus = []
with open('data4.txt', 'r', encoding="utf-8") as f:
    for line in f:
        corpus.append(" ".join(jieba.cut(line.split(',')[0], cut_all=False)))

vectorizer = TfidfVectorizer()
tfidf = vectorizer.fit_transform(corpus)
print (tfidf.shape)

words = vectorizer.get_feature_names()
for i in range(len(corpus)):
    print ('----Document %d----' % (i))
    for j in range(len(words)):
        if tfidf[i,j] > 1e-5:
              print (words[j], tfidf[i,j])
```

```python
jieba.set_dictionary("dict.txt.big.txt")
# 將關鍵字詞取出來 
# topK代表要取的關鍵字次數 
with open("data4.txt", "rb") as f:
    for line in f:
        tags = jieba.analyse.extract_tags(line,
                                          topK=10, 
                                          withWeight=True)
        for tag, weight in tags:
            print(tag + "," + str(int(weight * 10000)))
```

```python
#取每篇文章的 10 大關鍵字
with open("data4.txt", 'r', encoding="utf-8") as f:
    for line in f:
        tags = jieba.analyse.extract_tags(line,10)
        print(",".join(tags))
        
```
